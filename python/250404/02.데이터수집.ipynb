```python
# 환율 정보 csv 저장 자동화
file_name = "환율csv"
# 저장하려고 하는 주소에 file_name이 존재하는가?
header_bool = os.path.exists(file_name)
# file_name in os.listdir('./')
df = pd.read_html('https://finance.naver.com/marketindex/exchangeList.naver', encoding='cp949')[0]
df.to_csv(
    file_name,
    mode='a',
    index=False,
    header = header_bool
)
```

```python
import database
```

```python
# sql에 데이터를 대입
mydb = database.MyDB()
```

```python
insert_query = """
    insert into `exchange`(
        `통화명`, `매매기준율`, `현찰_사실때`, `현찰_파실때`, `송금_보내실때`, `송금_받으실때`, `미화환산율`)
        values(%s, %s, %s, %s, %s, %s, %s)
"""
mydb.execute_query(insert_query, *df.loc[0].values, inplace=True)
```

```python
df.fillna(0, inplace=True)
```

```python
# 데이터프레임의 길이만큼 반복 실행하는 반복문을 생성
for i in range(len(df)):
    # i -> 데이터프레임의 인덱스의 위치
    mydb.execute_query(insert_query, *df.loc[i].values, inplace=True)
```

```python
df.columns
```

```python
df.columns = df.columns.map(
    lambda x : x[0] if (x[0] == x[1]) else x[0] + '_' + x[1].replace(' ', '')
)
```

```python
from sqlalchemy import create_engine
```

```python
host = '127.0.0.1'
port = 3306
user = 'root'
pwd = '2069'
db = 'ubion'
```

```python
# Sqlalchemy에 엔진을 생성
engine = create_engine(f'mysql+pymysql://{user}:{pwd}@{host}:{port}/{db}')
```

```python
df.to_sql(
    name = 'exchange2',  # 테이블 이름
    con = engine,      # 데이터베이스 정보(sqlalcemy engine)
    if_exists = 'append',  # 테이블이 존재한다면 추가 옵션
    index = False
)
```

```python
# 몽고db에 데이터를 삽입
from pymongo import MongoClient
```

```python
df = pd.read_html('https://finance.naver.com/marketindex/exchangeList.naver', encoding='cp949')[0]
df.columns = df.columns.map(
    lambda x : x[0] if (x[0] == x[1]) else f"{x[0]}_{x[1].replace(' ', '')}"
)

data = df.to_dict(orient='records')
db['exchange2'].insert_many(data)
```
