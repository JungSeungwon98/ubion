## TextSplitter

```python
from langchain_community.document_loaders import PyPDFLoader

file_path = "/content/drive/MyDrive/금융빅데이터교재/RAG관련/nltk_data/SPRi AI Brief_4월호_산업동향_250407_F.pdf"
loader = PyPDFLoader(file_path)

# docs = loader.load()
docs = loader.load_and_split()
docs[1].page_content
```

```python
pages = docs[1].page_content
```

```python
documents = []
for i in docs:
  i.metadata["file_path"] = file_path
  # documents.append(i.page_content)
documents.extend(docs)
documents
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=20, length_function=len)
results = text_splitter.split_documents(documents)
results
```

```python
char_list = []
for i in range(len(results)):
  # print(len(results[i].page_content))
  char_list.append(len(results[i].page_content))

print(char_list)
```

## Text Embedding

```python
from langchain_openai import OpenAIEmbeddings
import os

os.environ["OPENAI_API_KEY"] = "openai_api_key를 입력하세요"
embeddings_model  = OpenAIEmbeddings(
    model="text-embedding-3-large",
    # With the `text-embedding-3` class
    # of models, you can specify the size
    # of the embeddings you want returned.
    # dimensions=1024,
)
embeddings = embeddings_model.embed_documents(
    [
        "안녕하세요",
        "제 이름은 홍길동입니다.",
        "이름이 무엇인가요?",
        "랭체인은 유용합니다.",
        "Hello World!"
    ]
)
len(embeddings), len(embeddings[0])
```

```python
from langchain_community.vectorstores import  FAISS
```

```python
vector_store = FAISS.from_documents(results, embedding=embeddings_model)
vector_store.save_local("faiss_index")
```

```python
# new_db = vector_store.load_local("faiss_index", embeddings_model)

query= "인공지능 산업구조는 어떻게 구성되어 있습니까?"
search_results = vector_store.similarity_search_with_relevance_scores(query=query , k=3)

for i in range(len(search_results)):
    print("{0}번째 유사 문서 유사도 \n{1}".format(i+1,round(search_results[i][1],2)))
    print("-"*100)
    print(search_results[i][0].page_content)
    print("\n")
    print(search_results[i][0].metadata)
    print("-"*100)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings

model_name = "jhgan/ko-sbert-nli"
model_kwargs = {'device': 'cpu'}
encode_kwargs = {'normalize_embeddings': True}
ko = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

```python
db = FAISS.from_documents(results, embedding=ko)
db.save_local("faiss_index")
```

```python
query= "인공지능 산업구조는 어떻게 구성되어 있습니까?"
search_results = db.similarity_search(query=query)
search_results[0]
```
